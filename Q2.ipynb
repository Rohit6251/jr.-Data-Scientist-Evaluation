```python
import language_tool_python
tool = language_tool_python.LanguageTool('en-US')
e=[]
for i in df1["text"]:
    m = tool.check(i)
    e.append(len(m)) 
```

```python
df1['Total mistakes']=e
```

```python
w=[]
for i in df1["text"]:
    m = tool.correct(i)
    w.append(m)
```

```python
df1['corrected mistakes']=w
```
